```python
#calculate accuracy of a test dataset per class
from keras.preprocessing import image
import glob
from keras.models import load_model
results=[]

correct=0
categorycount=0
for i in glob.glob("/...path.../chainsaw/*.jpg"):
  categorycount+=1
  n= image.load_img(i, target_size=(224,224))
  n = np.asarray(n)
  n = np.expand_dims(n, axis=0)
  output = saved_model.predict(n)
  if np.amax(output) == output[0][0] :
    correct += 1
results.append(correct/categorycount)

correct=0
categorycount=0
for i in glob.glob("/...path.../church/*.jpg"):
  categorycount+=1
  n= image.load_img(i, target_size=(224,224))
  n = np.asarray(n)
  n = np.expand_dims(n, axis=0)
  output = saved_model.predict(n)
  if np.amax(output) == output[0][1] :
    correct += 1
results.append(correct/categorycount)

correct=0
categorycount=0
for i in glob.glob("/...path.../dog/*.jpg"):
  categorycount+=1
  n= image.load_img(i, target_size=(224,224))
  n = np.asarray(n)
  n = np.expand_dims(n, axis=0)
  output = saved_model.predict(n)
  if np.amax(output) == output[0][2] :
    correct += 1
results.append(correct/categorycount)

correct=0
categorycount=0
for i in glob.glob("/...path.../fish/*.jpg"):
  categorycount+=1
  n= image.load_img(i, target_size=(224,224))
  n = np.asarray(n)
  n = np.expand_dims(n, axis=0)
  output = saved_model.predict(n)
  if np.amax(output) == output[0][3] :
    correct += 1
results.append(correct/categorycount)

correct=0
categorycount=0
for i in glob.glob("/...path.../frenchhorn/*.jpg"):
  categorycount+=1
  n= image.load_img(i, target_size=(224,224))
  n = np.asarray(n)
  n = np.expand_dims(n, axis=0)
  output = saved_model.predict(n)
  if np.amax(output) == output[0][4] :
    correct += 1
results.append(correct/categorycount)

correct=0
categorycount=0
for i in glob.glob("/...path.../golfball/*.jpg"):
  categorycount+=1
  n= image.load_img(i, target_size=(224,224))
  n = np.asarray(n)
  n = np.expand_dims(n, axis=0)
  output = saved_model.predict(n)
  if np.amax(output) == output[0][5] :
    correct += 1
results.append(correct/categorycount)

print(results)
#results shows the percentage of images correctly classified per class
```
